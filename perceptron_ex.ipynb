```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Create DataFrames

*   'or' dataframe
*   'and' dataframe
*   'xor' dataframe

```python
or_df = pd.DataFrame()
and_df = pd.DataFrame()
xor_df = pd.DataFrame()
```

```python
or_df['input1'] = [1,1,0,0]
or_df['input2'] = [1,0,1,0]
or_df['output'] = [1,1,1,0]
```

```python
and_df['input1'] = [1,1,0,0]
and_df['input2'] = [1,0,1,0]
and_df['output'] = [1,0,0,0]
```

```python
xor_df['input1'] = [1,1,0,0]
xor_df['input2'] = [1,0,1,0]
xor_df['output'] = [0,1,1,0]
```

### Output of each data frames and visual representation of dataset

```python
and_df
```

```python
sns.scatterplot(x='input1', y='input2', data=and_df, hue='output')
```

```python
or_df
```

```python
sns,sns.scatterplot(x='input1', y='input2', data=or_df, hue='output')
```

```python
xor_df
```

```python
sns.scatterplot(x='input1', y='input2', data=xor_df)
```

### Providing classifier and alocate dataset

```python
from sklearn.linear_model import Perceptron
clf1=Perceptron()
clf2=Perceptron()
clf3=Perceptron()
```

```python
clf1.fit(and_df.iloc[:,0:2].values, and_df.iloc[:,-1].values)
clf2.fit(or_df.iloc[:,0:2].values, or_df.iloc[:,-1].values)
clf3.fit(xor_df.iloc[:,0:2].values, xor_df.iloc[:,-1].values)
```

```python
clf1.coef_, clf1.intercept_
```

```python
x = np.linspace(-1,1,5)
y = -x+1
```

```python
sns.scatterplot(x='input1', y='input2', data=and_df, hue='output')
plt.plot(x,y)
```

```python
clf2.coef_, clf2.intercept_
```

```python
x1=np.linspace(-1,1,5)
y1=-x+0.5
sns.scatterplot(x='input1', y='input2', data=or_df, hue='output')
plt.plot(x1,y1)
```

```python
clf3.coef_, clf3.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(xor_df.iloc[:,0:2].values, xor_df.iloc[:,-1].values, clf3)
```

Conclused that perceptron is not able to classify 'xor' like problem.It can only work on linear problems. Its shows limitation of perceptron.
